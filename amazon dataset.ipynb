```python
#importing all useful libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
#load the csv file into jupyter notebook
amazon = pd.read_csv("Amazon_Products.csv")
```

```python
amazon.head()
```

```python
amazon.info()
#all columns are of 'object' datatype so we will have to convert those to numeric columns to perform anaylsis on them.
#.info() is not shoeing number if missing values in the dataset
```

```python
#to find the number of missing values in each attribute, we will use panadas isnull().sum()
amazon.isnull().sum()
```

```python
#there are many  columns with more than 60% of data missing so we wil drop these columns 
#dropping these columns using pandas dropna() and keeping thresh at 60%
amazon_new = amazon.dropna(axis = 1, thresh = 6000)
```

```python
amazon_new.head()
```

```python
amazon_new.info()
#columns are reduced from 895 to 13
#we can still see that there are missing values 
#first we will deal with unique ids 
```

```python
#2 unique ids are missing
#the only option here is to eliminate the rows with missing unique ids as it doesnt make any sense to replace it 
#with the median of the present values
#dropping the missing rows using .dropna()
amazon_new.dropna(subset=['uniq_id'], inplace = True)
```

```python
code_len= amazon_new["uniq_id"].map(lambda code: len(code))
code_len.unique()
```

```python
#we can see that there is/are unique ids which have only one character 
for i in amazon_new['uniq_id']:
    if len(i) == 1:
        print(i) 
```

```python
#we can see that there are two uniq ids which have '{' 
#so we will drop these rows 
amazon_new[amazon_new['uniq_id']=='}']
```

```python
#rows 2003 and 2005 has '}'
amazon_new.drop([2003,2005],inplace = True)
```

```python
amazon_new.info()
```

```python
amazon_new["price"]
```

```python
#the columns such as price, stock, reviews, answered questions, review rating are numeric columns 
#but these columns are displayed as object columns so we need to convert these to numeric columns 
#staring with the price column
amazon_new["price"]
```

```python
#TOOK REFRENCE FROM GOOGLE
#the price column has '£' sign which is resulting in it being idetifiied as an object 
#for items which have price greater than 999, these values might also contain a comma 
#removing the '£' and ','
# removing the pound sign
amazon_new["price_new"] = amazon_new["price"].str.split(" ", expand = True).get(0).str.split("£", expand = True).get(1)
# check if the value is in range of thousand
amazon_new[amazon_new.price_new.str.contains(',', na=False)]
# remove the comma sign
amazon_new["price_new"] = amazon_new["price_new"].str.replace(",", "")
```

```python
amazon_new["price_new"]
```

```python
amazon_new["price_new"] = amazon_new["price_new"].astype(float)
amazon_new["price_new"]
```

```python
amazon_new.head(5)
```

```python
#changing the data type of answered questions column to numeric
amazon_new["number_of_answered_questions_new"] = amazon_new["number_of_answered_questions"].astype(float)
```

```python
#removing ',' from the columns as it was giving an error
amazon_new["number_of_reviews_new"] = amazon_new["number_of_reviews"].str.replace(",","")
```

```python
#converting to numeric data type
amazon_new["number_of_reviews_val"] = amazon_new["number_of_reviews_new"].astype(float)
```

```python
#the number available in stocl contains "used" amd "new"
#so first we will have to seperate these 
amazon_new["number_available_in_stock"] = amazon_new["number_available_in_stock"].str.extract("(\d+)").astype(np.float)
```

```python
amazon_new["number_available_in_stock_new"] = amazon_new["number_available_in_stock"].astype(float)
amazon_new["number_available_in_stock"]
```

```python
amazon_new[["review", "out_of"]] = amazon_new["average_review_rating"].str.split(" out", expand = True)
```

```python
#converting review attribute to numeric value
amazon_new["review_new"] = amazon_new["review"].astype(np.float)
```

```python
amazon_new["number_of_reviews_new"] = amazon_new["number_of_reviews_new"].astype(float)
```

```python
amazon_new.info()
```

```python
#now dropping the columns which are of no more use to us 
amazon_new.drop([ "out_of","number_of_reviews_val","review","number_available_in_stock" ], axis = 1, inplace = True)
```

```python
#all numeric values which were of object type are converted to float 
#still there are missing values which will be dealt with later 
#the above part was data cleaning
#now exploratory data analysis 
values = amazon_new["manufacturer"].value_counts().keys().tolist()[:10]
values 
```

```python
#OUTSOURCED FROM GOOGLE 
df_list = []
for i in values:
    x = amazon_new[amazon_new["manufacturer"] == i]
    df_list.append(x)
frame = pd.concat(df_list)
frame.head()
```

```python
sns.catplot(data = frame, x = "manufacturer", kind = 'count')
```
